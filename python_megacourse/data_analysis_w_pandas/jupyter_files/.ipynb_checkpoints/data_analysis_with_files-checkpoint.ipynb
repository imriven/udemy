```python
df1=pandas.read_csv("https://pythonhow.com/supermarkets.csv")
df1
```

```python
# use drop method to delete columns or rows from a data frame (column or row, how many)
# if you want to update the data frame permanently without the column or row you have to reassign the variable
# df1= df1.drop("City",1)
df1.drop("City",1)
```

```python
#you can also drop rows and columns by indexing

df1.drop(df1.index[0:3],0)
```

```python
#you can also index columns

df1.drop(df1.columns[0:3],1)
```
